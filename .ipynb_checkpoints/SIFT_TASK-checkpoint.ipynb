```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Helper function to display images in Jupyter Notebook
def display_image(title, img):
    if img is None:
        print("Error: Image not found. Please check the file path.")
        return  # Exit the function if the image isn't loaded

    plt.figure(figsize=(10, 5))
    plt.imshow(img,)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the image and check for success
image_path = './images/sample_image.jpg'
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Unable to load image at path {image_path}")
else:
    # Display the image if successfully loaded
    display_image('Loaded Image', image)
```
